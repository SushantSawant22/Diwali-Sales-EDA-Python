## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Importing Dataset

```python
dataset=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape') #To avoid encoding error use unicode_escape
dataset
```

```python
print(dataset.shape)
print(dataset.info())
```

## Data Cleaning

```python
dataset.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

```python
dataset
```

```python
dataset.info()
```

```python
dataset['Amount']=dataset['Amount'].astype('int')
```

```python
dataset
```

```python
dataset.rename(columns={'Marital_Status':'Marriage'},inplace=True)
```

```python
dataset[['Age','Orders','Amount']].describe() #describe() method returns statistical information
```

## Exploratory Data Analysis

```python
dataset
```

## Gender

```python
dataset['Gender'].replace('F','Female',inplace=True)
```

```python
dataset['Gender'].replace('M','Male',inplace=True)
```

```python
ax=sns.countplot(x='Gender',data=dataset)
sns.set(rc={'figure.figsize':(3,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Females placed more Orders than Male

```python
sales_gender=dataset.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_gender
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gender)
sns.set(rc={'figure.figsize':(3,3)})
```

### Female has more purchasing power than Male

## Age Group

```python
ax=sns.countplot(data=dataset,x='Age Group',hue='Gender')
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Most Number of Orders were placed by Age Group 26-35 years 

```python
sales_age=dataset.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Age Group',ascending=False)
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

### Age Group 26-35 years has more Purchasing Power 

## State

```python
sales_state=dataset.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

### Utter Pradesh and Maharashtra has placed most number of orders 

```python
sales_state=dataset.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

### Uttar Pradesh and Maharshtra has highest Purchasing Power 

## Marriage 

```python
dataset['Marriage'].replace(0,'Single',inplace=True)
dataset['Marriage'].replace(1,'Married',inplace=True)
```

```python
dataset
```

```python
ax=sns.countplot(x='Marriage',data=dataset)
sns.set(rc={'figure.figsize':(3,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Single People had placed more number of orders 

```python
sales_marriage=dataset.groupby(['Marriage','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_marriage
```

```python
sns.barplot(x='Marriage',y='Amount',data=sales_marriage,hue='Gender')
sns.set(rc={'figure.figsize':(7,5)})
```

### Among Single People Female has most purchasing power

## Occupation 

```python
ax=sns.countplot(data=dataset,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### People from IT Sector, Healthcare and Aviation placed most orders

```python
sales_ocp=dataset.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
```

```python
sns.barplot(x='Occupation',y='Amount',data=sales_ocp)
sns.set(rc={'figure.figsize':(7,5)})
```

### IT sector, Healthcare, Aviation are some occupations having most purchasing power 

## Product 

```python
ax=sns.countplot(data=dataset,x='Product_Category')
sns.set(rc={'figure.figsize':(35,10)})
for bars in ax.containers:
    ax.bar_label(bars)
```

### Food and Clothing Products are ordered most by customers 

```python
sales_prod=dataset.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(12)
sales_prod
```

```python
palette_color = sns.color_palette('bright')
plt.pie(sales_prod['Amount'],labels=sales_prod['Product_Category'],colors=palette_color, autopct='%.0f%%')
  
plt.show()
```

### Most sale is done by Food product, Food and Clothing,Apparel products make approx 50% profit

Unmarried women from age group 26-35 years of UP, Maharashtra from IT Sector and Aviation Industry are more likely to buy Products from Food,Clothing and Electronics Category
